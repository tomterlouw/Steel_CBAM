### Make beautiful Sankey diagrams for the steel production pathways

```python
databases = [NAME_REF_DB, 'ecoinvent_image_SSP2-L_2040_base', 'ecoinvent_remind_SSP2-PkBudg1000_2040_base']

for db_sel in databases:
    steel_activities = [
            act for act in bd.Database(db_sel)
            if ", with transport" not in act["name"] and act['name'] != "steel production, converter, unalloyed, only primary steel"
            and (
                ( ("steel production, " in act["name"]) and (act["location"] == "GLO") and (act["reference product"] == "steel, low-alloyed") )
                or (
                    "primary steel, using direct reduced iron" in act["name"]
                    and act["location"] in ["RoW", "World"]
                )
                or (
                    "steel production, electric, low-alloyed, secondary steel, using scrap iron" in act["name"]
                    and act["location"] in ["RoW", "World"]
                )
            )
        ]
    
    for act in steel_activities:
        print(act['name'], act['location'])
    
        sankey(
            activity=act,
            level=5,
            cutoff=0.01,
            method=MY_METHODS[0],
            notebook=False,
            filepath=f"figs/sankeys/{db_sel} {act['name'][:30]}.html"
        )
```
