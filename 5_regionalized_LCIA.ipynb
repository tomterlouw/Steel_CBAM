```python
from edges import get_available_methods
from edges import EdgeLCIA
import bw2data as bd
from config import *

# Get the available methods
methods = get_available_methods()
list(methods)
```

```python
# Import project
bd.projects.set_current(PROJECT_NAME) #Creating/accessing the project
DB_NAME_INIT, bd.databases,PROJECT_NAME
```

```python
# Select an activity from the LCA database
act = [act for act in bd.Database("steel_market_db") if act['location'] == "GLO"][0]

# Define a method
method = ('AWARE 2.0', 'Country', 'all', 'yearly') # one could even go into more detail, more precise based on lon, lat (since we know the province/watershed)
#method = ('IBIF', 'biodiversity', 'all pressures', 'overall') # one could even go into more detail, more precise based on lon, lat (since we know the province/watershed)


# Initialize the LCA object
LCA = EdgeLCIA({act: 1}#1918*1e6
               , method) #Mt of steel, but the amount does not matter for the mapping and CF evaluation steps
LCA.lci()

# Map CFs to exchanges: apply suggested strategies
LCA.apply_strategies()

# or apply these strategies manually
#LCA.map_exchanges()

# If needed, extend the mapping to aggregated and `dynamic` regions (e.g., RoW)
LCA.map_aggregate_locations()
LCA.map_dynamic_locations()
LCA.map_contained_locations()

# add global CFs to exchanges missing a CF
LCA.map_remaining_locations_to_global()

# Evaluate CFs
LCA.evaluate_cfs()

# Perform the LCIA calculation
LCA.lcia()
print(LCA.score)

# Optional but recommended: print a dataframe with the characterization factors used
# this allows you to check whether exchanges have been given the correct CFs
# include_unmatched=True allows you to see which exchanges were not matched (and if some should have been)
df = LCA.generate_cf_table(split_aggregate_consumers=True)
df.to_excel("cf_table.xlsx", index=False) # each unique activity, determine the net water use per province/country.
```
